```python
import pandas as pd
options=webdriver.ChromeOptions()
    #run Selenium in headless mode
options.add_argument('--headless')
options.add_argument('--no-sandbox')
    #overcome limited resource problems
options.add_argument('--disable-dev-shm-usage')
    #open Browser in maximized mode
options.add_argument("start-maximized")
    #disable infobars
options.add_argument("disable-infobars")
    #disable extension
options.add_argument("--disable-extensions")
options.add_argument("--incognito")
options.add_argument("--disable-blink-features=AutomationControlled")
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

#Variable du nombre de page
page_number = list(range(1,5+1))

driver = webdriver.Chrome('chomedriver', options=options)

### SCRAPPING DE LA PAGE DES HOMMES ###

#Création du dictionnaire de stockage des valeures récupérée
data_dict = {"articles_nom":[], "article_detail":[], "articles_origine":[],"articles_composition":[], "genre":[]};

list_link_products = []

#BOUCLE D'INCRÉMENTATION DES PAGES 
for i in page_number:
    
    Page_homme=f'https://www.cocorico.store/183-tout-voir?p={i}'
    driver.get(Page_homme)
    
    #Récupération des liens des produits
    products = driver.find_elements(By.XPATH, "//*[@class='product-bottom-card animall']/a")
    
    #BOUCLE DE RECUPÉRATION DES LIENS SUR CHAQUE PRODUIT
    for i in range(0,len(products)):
        #try:
            link_product = products[i].get_attribute('href')
            list_link_products.append(link_product)
        #except:
            #pass    
    list_link_products = list(set(list_link_products))

     #BOUCLE DE PARCOURS DE LIEN DE CHAQUE PRODUIT
    for link in list_link_products:
        
        #Ici, on met a jour notre driver pour qu'il charge une nouvelle page en f(x) du lien donné
        driver.get(link)
        
        article_nom = driver.find_elements(By.XPATH, "//*[@id='h1_complement']/p") 
        
        article_detail = driver.find_elements(By.XPATH, "//*[@itemprop='name']")
           
        article_origine = driver.find_elements(By.XPATH, "//*[@class='origine']")
        
        #//*[@id="collapse-product-entretien"]/div/p
        #/html/body/div[1]/div[2]/div/div[3]/div/div/div[1]/div[3]/div[4]/div[2]/div[2]/div/p/text()[2]
        boutton_click = driver.find_element(By.XPATH, "//*[@id='product-entretien-header']/button").click()
        
        #time.sleep(5)
        #boutton_click.click()
                                 
        article_composition = driver.find_elements(By.XPATH, "//p[strong = 'Matière']")
        
        #//*[@id="collapse-product-entretien"]/div/p/text()[2]
        #//*[@id="collapse-product-entretien"]/div/p/text()[2]
        #&nbsp;: 100% Coton
        #/html/body/div[1]/div[2]/div/div[3]/div/div/div[1]/div[3]/div[4]/div[2]/div[2]/div/p/text()[2]
        #article_composition = driver.find_elements(By.XPATH, "//*[@id='collapse-product-entretien']/div/p")
        
        #ALIMENTATION DU DICTIONNAIRE AVEC LES VALEURES RÉCUPÉRÉES LOSQU'ON EXPLORE UN LIEN        
        for i in range(len(article_nom)):   
            
            try:
                article=article_nom[i].text.strip()
                data_dict["articles_nom"].append(article)
            except IndexError:
                data_dict["articles_nom"].append("No Article Name")
                pass
            
            try:
                article=article_detail[i].text.strip()
                print(article)
                data_dict["article_detail"].append(article)
            except IndexError:
                data_dict["article_detail"].append("No Article Detail")
                pass
            
            try:
                article=article_origine[i].text.strip()
                data_dict["articles_origine"].append(article)
            except IndexError:
                data_dict["articles_origine"].append("No Article Origin")
                pass
            
            try:
                article=article_composition[i].text.strip()
                #article = article.split(":")[-1].strip()
                print(article)
                data_dict["articles_composition"].append(article)
            except IndexError:
                data_dict["articles_composition"].append("No Article Composition")
                print("Ca passe pas ")
                pass
            
            data_dict["genre"].append("Homme")



```

```python
dataFrameHomme = pd.DataFrame(data_dict)
dataFrameHomme.head(56)
```
