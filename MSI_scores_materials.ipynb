```python
import json
import pandas as pd

# Charger les données JSON à partir d'un fichier
with open('materials_msi_scores.json', 'r') as file:
    data = json.load(file)

# Extraire les données pertinentes
materials_list = []

for material in data['hits']:
    material_name = material['name']
    material_impact = material['scores']

    materials_list.append({
        'name': material_name,
        'globalWarmingPts': material_impact['globalWarmingPts'],
        'eutrophicationPts': material_impact['eutrophicationPts'],
        'waterScarcityPts': material_impact['waterScarcityPts'],
        'abioticDepletionPts': material_impact['abioticDepletionPts'],
        'chemistryPts': material_impact['chemistryPts'],
        'impact': material_impact['impact'],
        'waterConsumption': material_impact['waterConsumption'],
        'biogenicCarbon': material_impact['biogenicCarbon']
    })

# Créer un DataFrame pandas
df_materials_score_msi = pd.DataFrame(materials_list)

# Exporter les données dans un fichier CSV
df_materials_score_msi.to_csv("higgs_score_materials.csv", index=False)
df_materials_score_msi.head(25)
```
